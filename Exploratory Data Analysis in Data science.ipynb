## Visualization of EPL Dataset

```python
plt.hist(df["Age"])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.distplot(df['Age'], hist=True, color='red')
plt.xlabel("Ages", fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title("Distribution of Ages", fontsize=16)
```

```python
sns.scatterplot(x="Goals", y="Assists", data=df)
plt.show()
```

```python
plt.figure(figsize=(13, 7))
sns.scatterplot(x="Age", y="Mins", data=df)
plt.xlabel("Age")
plt.ylabel("Minutes Played")
plt.title("Age verses Minutes Played", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(13, 7))
sns.scatterplot(x="Matches", y="Mins", data=df)
plt.xlabel("Matches")
plt.ylabel("Minutes Played")
plt.title("Matches verses Minutes Played", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(13, 7))
sns.scatterplot(x="Red_Cards", y="Assists", data=df)
plt.xlabel("Red_Cards")
plt.ylabel("Assist")
plt.title("Red Card verses Assist", fontsize=20)
plt.show()
```

```python
# calculate correlation matrix
fig, ax = plt.subplots(figsize=(9,9))  
corr = df.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True), ax=ax)
```

```python
v = ['Goals']
for colname in v:
    x = x=pd.DataFrame(df.groupby(['Name','Club'])[colname].sum().reset_index(drop=False).sort_values(by=colname,ascending=True).tail(20).reset_index(drop=True))
    display(x.style.background_gradient( cmap='ocean'))
    sns.set(rc={'figure.figsize':(16,6.5)})
    sns.barplot(x='Name',y=colname,data=x,palette="ocean",order=x['Name'].tail(20))
    plt.title('The most {} in the EPL 2020 - 2021'.format(colname),fontsize=20)
    plt.ylabel('count of {}'.format(colname),fontsize=18)
    plt.xlabel(' ')
    plt.xticks(ticks=np.arange(0,20,1),labels=x['Name'],fontsize=16,style='oblique',rotation=90)
    plt.yticks(fontsize=16)
    plt.show()
```

```python
v = ['Yellow_Cards']
for colname in v:
    x = x=pd.DataFrame(df.groupby(['Name','Club'])[colname].sum().reset_index(drop=False).sort_values(by=colname,ascending=True).tail(20).reset_index(drop=True))
    display(x.style.background_gradient( cmap='ocean'))
    sns.set(rc={'figure.figsize':(20,7.5)})
    sns.barplot(x='Name',y=colname,data=x,palette="ocean",order=x['Name'].tail(20))
    plt.title('The most {} in the EPL 2020 - 2021'.format(colname),fontsize=20)
    plt.ylabel('count of {}'.format(colname),fontsize=18)
    plt.xlabel(' ')
    plt.xticks(ticks=np.arange(0,20,1),labels=x['Name'],fontsize=18,style='oblique',rotation=90)
    plt.yticks(fontsize=18)
    plt.show()
```
